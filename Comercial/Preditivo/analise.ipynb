```python
invalido = set()
```

```python
def replace_job_titles(cargo):
    for atalho, substituto in job_title_replacements:
        if re.search(atalho, cargo, re.IGNORECASE):
            return substituto
    invalido.add(cargo)
    return 'Invalido'
```

```python
df['Cargo'] = df['Cargo'].fillna('Vazio').apply(replace_job_titles)
```

```python
grouped = df.groupby(['Cargo', 'STATUS']).size().reset_index(name='Count')
```

```python
def rank_by_status(status):
    return grouped[grouped['STATUS'] == status].sort_values(by='Count', ascending=False)

ranked_projeto = rank_by_status('Projeto')

ranked_lead = rank_by_status('Lead')

ranked_reuniao_marcada = rank_by_status('Reunião Marcada')
```

```python
print("Ranking by 'Projeto':")
print(ranked_projeto)
```

```python
print("Ranking by 'Lead':")
print(ranked_lead)
```

```python
print("Ranking by 'Reuniao Marcada':")
print(ranked_reuniao_marcada)
```
