Selecao e analise de variavel por arquivo:

```python
import pandas as pd
import numpy as np
```

```python
def process_files_in_folder(atalho_pasta):
    for filename in os.listdir(atalho_pasta):
        atalho_arquivo = os.path.join(atalho_pasta, filename)
        df = pd.read_csv(atalho_arquivo) 

        print(f"Colunas em {filename}:")

        for col_name in df.columns:
            if not df[col_name].isna().all():  
                non_nan_value = df.loc[df[col_name].first_valid_index(), col_name]  
                print(f"{col_name}: {non_nan_value}")
            else:
                print(f"{col_name}: NaN")
        print("\n")
```

Altera nome das colunas para seus significados mapeados

```python
def process_table(csv_file, columns_dict, colunas_significados):
    df = pd.read_csv(csv_file)

    relevant_columns = [col for col in df.columns if col in columns_dict]
    df = df[relevant_columns]

    df.rename(columns=colunas_significados, inplace=True)

    df.to_csv("novo-" + csv_file, index=False)
```
