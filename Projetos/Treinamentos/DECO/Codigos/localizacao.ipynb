```python
df_coordinates['NOME_MUNICIPIO'] = df_coordinates['NOME_MUNICIPIO'].apply(lambda x: unidecode(str(x).lower()))
df['customer_city'] = df['customer_city'].apply(lambda x: unidecode(str(x).lower()))
df['seller_city'] = df['seller_city'].apply(lambda x: unidecode(str(x).lower()))

def get_coordinates(city_name, df_IBGE):
    city_row = df_IBGE[df_IBGE['NOME_MUNICIPIO'] == city_name]
    if city_row.empty:
        return None
    latitude = city_row.iloc[0]['LATITUDE'].replace(',', '.')
    longitude = city_row.iloc[0]['LONGITUDE'].replace(',', '.')
    return (latitude, longitude)

for index, row in df.iterrows():
    coords_cliente = get_coordinates(row['customer_city'], df_coordinates)
    coords_vendedor = get_coordinates(row['seller_city'], df_coordinates)
    
    if coords_cliente and coords_vendedor:
        df.at[index, 'distance'] = geodesic(coords_cliente, coords_vendedor).km

df_with_distance = df[df['distance'] > 0]
df_no_movement = df[df['distance'] == 0]

df_with_distance.to_csv(r"C:\Users\gabri\Documents\PROJETOS\PY\PJ_Code\DE\Data\Modelo 2\dados_completos.csv", index=False)
df_no_movement.to_csv(r"C:\Users\gabri\Documents\PROJETOS\PY\PJ_Code\DE\Data\Distancias\cidadesInvalidas.csv", index=False)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.distance import distance
import requests
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderQueryError, GeocoderServiceError

def simplify_address(address_data):
    # Initialize an empty list to hold address components
    address_parts = []

    # Add available address components in order of specificity
    if 'logradouro' in address_data and address_data['logradouro']:
        address_parts.append(address_data['logradouro'])
    if 'bairro' in address_data and address_data['bairro']:
        address_parts.append(address_data['bairro'])
    if 'localidade' in address_data and address_data['localidade']:
        address_parts.append(address_data['localidade'])
    if 'uf' in address_data and address_data['uf']:
        address_parts.append(address_data['uf'])

    # Always append 'Brazil' to ensure Geopy understands the country context
    address_parts.append('Brazil')

    # Construct and return the simplified address string
    simplified_address = ', '.join(address_parts)
    return simplified_address if simplified_address else None

def getAdress(cep):
    global InvalidCEP
    try:
        response = requests.get(f"https://viacep.com.br/ws/{cep}/json/")
        if response.status_code == 200:
            address_data = response.json()
            simplified_address = simplify_address(address_data)
            if simplified_address:
                return simplified_address
            else:
                print(f"falha no cep {cep}")
                return None
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
        # Handle retries or failure as appropriate
        InvalidCEP = True  # Assuming InvalidCEP is used to track failed lookups
        return None
    
def getCoordinates(address):
    # Basic validation of the address string
    if not address or not isinstance(address, str):
        print("Invalid address provided.")
        return None

    try:
        # Attempt to geocode using the provided address string
        location = geolocator.geocode(address)
        if location:
            # If geolocation is successful, return latitude and longitude as a tuple
            print('sucesso')
            return (location.latitude, location.longitude)
        else:
            # If geolocation fails (e.g., address not found), print an error and return None
            print(f"Geolocation failed for address: {address}")
            return None
    except GeocoderQueryError as e:
        print(f"Query error occurred: {e}")
        return None
    except GeocoderServiceError as e:
        print(f"Service error occurred: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

DATA_URL = r"C:\Users\gabri\Documents\PROJETOS\PY\PJ_Code\DE\Data\dados_distance_zero.csv"
df = pd.read_csv(DATA_URL)

# configura o geolocator
geolocator = Nominatim(user_agent="geolocalização")

# lista (coluna) das distancias
distanciasList = []

for cepCliente, cepVendedor in zip(df['customer_zip_code_prefix'], df['seller_zip_code_prefix']):
    # preenche o prefixo do cep com zeros a direita ate ter 8 digitos
    cepCliente = str(cepCliente).ljust(8, '0')
    cepVendedor = str(cepVendedor).ljust(8, '0')
    
    #tenta obter endereco pelo CEP
    enderecoCliente = getAdress(cepCliente)
    enderecoVendedor = getAdress(cepVendedor)
    
    # obtem distancia a partir do endereco
    if enderecoCliente and enderecoVendedor:
        coordenadasCustomer = getCoordinates(enderecoCliente) 
        coordenadasSeller = getCoordinates(enderecoVendedor)
        
        if coordenadasCustomer and coordenadasSeller:
            df['distance'] = distance(coordenadasCustomer, coordenadasSeller).km
        else:
            df['distance'] = 0
            print('falha na distancia')
            
            
data_with_distance = df[df['distance'] != 0]
data_without_distance = df[df['distance'] == 0]

# File paths
complete_data_path = r'C:\Users\gabri\Documents\PROJETOS\PY\PJ_Code\DE\Data\Modelo 2\dados_completos.csv'
new_data_path = r'C:\Users\gabri\Documents\PROJETOS\PY\PJ_Code\DE\Data\Distancias\dados_distance_zero.csv'

# Append data with distance to the existing file
data_with_distance.to_csv(complete_data_path, mode='a', header=False, index=False)

# Save data without distance to a new file
data_without_distance.to_csv(new_data_path, index=False)
```
